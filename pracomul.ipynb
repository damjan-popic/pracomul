```python
# ════ Cell 1 – Load data & pre‑process ════
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

DATA_DIR = Path("results")          # tweak if your folder differs

tok = pd.read_csv(
    DATA_DIR / "tokens_with_meta.csv",
    dtype={
        "token":         "string",
        "lemma":         "string",
        "university":    "category",
        "gender":        "category",
        "mother_tongue": "category",
        "spanish_level": "category",
    }
)

# Discourse‑marker list (one per line)
phrases = """
a lo mejor
a ver
además
afortunadamente
al fin y al cabo
al menos
al mismo tiempo
apenas
aun así
básicamente
bien
bueno
claramente
claro
completamente
concretamente
de acuerdo
de alguna forma
de alguna manera
de hecho
de todas formas
de todos modos
de verdad
definitivamente
desde luego
digamos
efectivamente
en cambio
en ese sentido
en fin
en general
en particular
en realidad
en verdad
entonces
es decir
eso sí
especialmente
exactamente
exacto
fíjate
generalmente
hasta
hombre
igual
incluso
justamente
justo
la verdad
más bien
más o menos
menos mal
mira
nada
ni siquiera
¿no?
no sé
o sea
obviamente
oye
perdón
perdona
pero
por (lo) tanto
por cierto
por ejemplo
por el contrario
por eso
por lo menos
por supuesto
por un lado
por otro lado
por una parte
por otra parte
porque
prácticamente
precisamente
pues
pues bien
pues nada
pues sí
realmente
seguramente
seguro
sin embargo
sobre todo
solamente
solo
también
tampoco
total
totalmente
vale
vamos
vaya
(y) es que
ya
""".strip().lower().splitlines()

# Lower‑cased token column for matching
tok["token_lc"] = tok["token"].str.lower()
```

```python
# ════ Cell 2 – Pivot + simple bar chart ════
# 1) Aggregate raw counts
acc = (
    tok.loc[tok["token_lc"].isin(phrases)]
       .groupby(["token_lc", "spanish_level", "university"], observed=True)
       .size()
       .rename("n")
       .reset_index()
)

# 2) CEFR bucket
upper_levels = {"B2", "C1", "C2"}
acc["bucket"] = np.where(acc["spanish_level"].isin(upper_levels), "upper", "lower")

# 3) Pivot to (token, university) → counts, then add total
pivot = (
    acc.pivot_table(index=["token_lc", "university"],
                    columns="bucket", values="n",
                    fill_value=0, observed=True)
       .reset_index()
)
pivot["freq_total"] = pivot["upper"] + pivot["lower"]

# 4) Pick the 15 most frequent phrases overall
top_phrases = (
    pivot.groupby("token_lc", observed=True)["freq_total"]
         .sum()
         .nlargest(15)
         .index
)

bar_data = (
    pivot[pivot["token_lc"].isin(top_phrases)]
        .groupby("token_lc", observed=True)[["lower", "upper"]]
        .sum()
)

# 5) Sort by total so the highest bars appear at top
bar_data["total"] = bar_data["lower"] + bar_data["upper"]
bar_data = bar_data.sort_values("total")

# 6) Plot: stacked horizontal bars
fig, ax = plt.subplots(figsize=(8, 7))
ax.barh(bar_data.index, bar_data["lower"], color="#1f77b4", label="lower (A1–B1)")
ax.barh(bar_data.index, bar_data["upper"],
        left=bar_data["lower"], color="#ff7f0e", label="upper (B2–C2)")

# Labels & legend
ax.set_xlabel("Raw token count")
ax.set_title("Top 15 discourse markers across all universities\n"
             "(stacked counts by CEFR bucket)")
ax.legend(title="CEFR bucket", loc="lower right")
ax.grid(axis="x", ls="--", alpha=.3)

# Optional: annotate counts at bar ends
for y, (low, up) in enumerate(zip(bar_data["lower"], bar_data["upper"])):
    ax.text(low / 2, y, str(low), va="center", ha="center", color="white", fontsize=8)
    ax.text(low + up / 2, y, str(up), va="center", ha="center", color="black", fontsize=8)

plt.tight_layout()
plt.show()
```
